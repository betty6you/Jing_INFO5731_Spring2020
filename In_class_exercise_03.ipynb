```python
# You code here (Please add comments in the code):
# This code is created to solve the first step in Question 1
import urllib.request
from bs4 import BeautifulSoup

mybaseURL="https://pubmed.ncbi.nlm.nih.gov/?term=chinese%20medicine%20epilepsy&page="
starURL=[]
for i in range(1,51):
  starURL=mybaseURL+str(i)

  with urllib.request.urlopen(starURL) as f:
    webpage=f.read()
    soup=BeautifulSoup(webpage)
  #print(soup)
  for item in soup.select('[data-rel-pos]'):
  
    print('---------------------------------------------------------------------------------------------------------------------')
  #print(item)
    print(item.select('a.docsum-title')[0].get_text()) # The title of the article
    print(item.select('a.docsum-title')[0]['data-article-id']) # The PMID of the article
        

```
