```
# Write your code here
!pip install pandas
import pandas as pd
import string
import nltk
from nltk.util import ngrams
nltk.download('punkt')
data=pd.read_csv(r"https://raw.githubusercontent.com/betty6you/Jing_INFO5731_Spring2020/main/AmazonReviewupdated.csv")

# load cleaned and preprocessed text from assignment 2
f=open('AmazonReviewupdated.csv')
text=f.read()
tokens = nltk.word_tokenize(text)

#Creating trigrams
trigrams = nltk.trigrams(tokens)

#computing frequency distribution for all the trigrams in the text
freq_dist = nltk.FreqDist(trigrams)
for k,v in freq_dist.items():
    print(k,v)
```
